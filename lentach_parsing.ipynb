```python
import vk
import time
from IPython.display import clear_output
import nltk
from nltk.stem import WordNetLemmatizer
import operator
import pymorphy2
```

```python
def get_wall(sleep_time=0.3, all_history=None, old_friends=None, stop=None):
    i = 0
    if all_history is None:
        all_history={}

    time.sleep(sleep_time)
    friend_history = []
    clear_output(wait=True)
    if all_count > 200:
        resid = all_count
        offset = 0
        while resid > 0:
            success = 0
            while success==0:
                try:
                    friend_history += vkapi('wall.get', owner_id=-29534144, filter='owner', count=100, offset=offset)
                    time.sleep(sleep_time)
                    success=1
                except:
                    success=0

            resid -= 100
            offset += 100
            if resid > 0:
                print('--processing', resid, 
                    'of', all_count, 'messages left')
            if stop is not None:
                if stop:
                    break
        all_history = friend_history
    return all_history
```

```python
texts = []
for history in result[1:]:
    print(str(type(history)))
    if str(type(history))!='<class \'int\'>':
        if 'text' in history:
            texts.append(history['text'])
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
tokens = nltk.word_tokenize(' '.join(texts))
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
lemmatized_tokens = [morph.parse(token.lower())[0].normal_form for token in tokens]
```

```python
lemmatized_tokens
```

```python
freq_dict = {}
for token in lemmatized_tokens:
    if token not in freq_dict.keys():
        freq_dict[token]=1
    else:
        freq_dict[token]+=1
```
