```python
import pandas as pd
import cv2
import urllib.request
import os
from tqdm import tqdm_notebook
from utils import sentence_processor
from IPython.display import clear_output
```

```python
# Read dataset
dataset_raw = pd.read_csv('Dataset/dataset_raw.csv')
dataset_raw = dataset_raw.rename(index=str, columns={'Unnamed: 0':'id'})
dataset_raw.head()
```

```python
# download  photos
photos = dataset_raw.photos
ids = dataset_raw.id
for id,photo in zip(ids[0:200],photos[0:200]):
    clear_output(wait=True)
    print(id)
    
    success = 0
    try_num = 0
    while success==0:
        try:
            if photo != 'no_photo':
                urllib.request.urlretrieve(photo, 'Dataset/images/'+str(id)+'.jpg')
            success =1
            try_num=0
        except:
            success=0
            try_num +=1
        if try_num == 10:
            success =1
```
